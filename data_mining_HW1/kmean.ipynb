```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
df = pd.read_csv('201703_Taiwan.csv')
df = df[df['device_id']=='74DA388FF5F6']
df = df.reset_index(drop=True)
df = df[['PM1','PM10','Humidity']]
X = df.as_matrix()

kmeans = KMeans(n_clusters=15, random_state=0).fit(X)
kmeans.labels_

for i in range(0,len(df)):
    tmp = kmeans.labels_[i] #0~9
    df['PM1'][i] = kmeans.cluster_centers_[tmp][0]
    df['PM10'][i] = kmeans.cluster_centers_[tmp][1]
    df['Humidity'][i] = kmeans.cluster_centers_[tmp][2]


print(kmeans.cluster_centers_)
print(df)
for j in df.index:
    print('pm1:',df['PM1'][j],' pm10:',df['PM10'][j],' humidity:',df['Humidity'][j],sep='')
```
